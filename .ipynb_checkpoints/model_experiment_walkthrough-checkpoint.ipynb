# 🚀 Walkthrough Lengkap: Prediksi Workload & Auto-Scaling
Dokumen ini merupakan implementasi end-to-end (Pipeline) dari pemrosesan metrik kluster hingga pemodelan *Deep Learning* tingkat lanjut untuk keperluan *Container Auto-Scaling*.
Model yang akan dibangun, dibandingkan, dan dievaluasi meliputi: **LSTM**, **GRU**, **BiLSTM**, dan **DUCFF (TCN + Transformer)**.
Target prediksi kita pada walkthrough ini adalah **cpu_media** (beban CPU), yang nantinya digunakan untuk menentukan strategi pembatasan CPU.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, LSTM, GRU, Bidirectional, Dropout, LayerNormalization
from tensorflow.keras.callbacks import EarlyStopping

import warnings
warnings.filterwarnings('ignore')
```

## 2.3 Preprocessing / Pembersihan Data
Data mentah jarang siap pakai. Tahap ini meliputi: 
- Menangani data hilang (*missing values*), 
- Menghapus duplikasi, 
- Menangani *outlier*, 
- *Encoding* data kategorikal (jika ada),
- Serta **normalisasi/standardisasi** skala data agar seluruh fitur berada pada rentang yang sebanding (sangat krusial untuk Neural Network).


```
# 1. Load Data
df = pd.read_csv('all_merged_metrics.csv')

# 2. Mengubah timestamp menjadi format Datetime (Asumsi timestamp adalah Epoch seconds)
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df.set_index('timestamp', inplace=True)
print("Data awal shape:", df.shape)

# 3. Menghapus Duplikasi
df.drop_duplicates(inplace=True)

# 4. Menangani Missing Values (Menggunakan interpolasi waktu untuk data deret waktu)
df.interpolate(method='time', inplace=True)
df.bfill(inplace=True) # Mengisi sisa missing value di awal data jika ada

# 5. Menangani Outlier pada 'cpu_media' menggunakan metode Z-Score yang di-clip (agar tidak menghapus terlalu banyak baris data)
Q1 = df['cpu_media'].quantile(0.25)
Q3 = df['cpu_media'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
df['cpu_media'] = np.where(df['cpu_media'] > upper_bound, upper_bound, df['cpu_media'])

# 6. Normalisasi Data (Scaling ke rentang 0 - 1)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled_data, columns=df.columns, index=df.index)

print("Data bersih shape:", df_scaled.shape)
df_scaled.head()
```

## 2.4 Feature Engineering & Seleksi Fitur
Membuat fitur baru yang lebih informatif dari data yang ada, atau memilih fitur yang paling relevan dan membuang fitur yang tidak berpengaruh, agar model belajar lebih efisien dan tidak 'bingung' oleh data yang tidak relevan.

Karena kita bekerja dengan Time-Series, fitur terpenting adalah **Lag Features** (Data historis t-1, t-2, ... t-N) untuk memprediksi masa depan (t+1).


```
# Seleksi Fitur (Menggunakan fitur yang berkorelasi tinggi dengan beban CPU)
# Dalam konteks auto-scaling, CPU, RAM, dan RPS saling berkaitan.
features = ['cpu_media', 'rps_media', 'ram_media']
df_features = df_scaled[features]

# Fungsi untuk membuat lag features (sliding window)
def create_dataset(dataset, time_steps=1):
    X, Y = [], []
    for i in range(len(dataset) - time_steps):
        # Ambil jendela waktu (window) sebesar time_steps
        X.append(dataset[i:(i + time_steps), :])
        # Y adalah target pada titik t+1 (kolom indeks 0 adalah 'cpu_media')
        Y.append(dataset[i + time_steps, 0])
    return np.array(X), np.array(Y)

TIME_STEPS = 5  # Menggunakan 5 step sebelumnya untuk memprediksi step selanjutnya
X, y = create_dataset(df_features.values, TIME_STEPS)

print("Shape X (Input) :", X.shape) # (Samples, Time Steps, Features)
print("Shape y (Target):", y.shape) # (Samples, )
```

## 2.5 Pemilihan Model (Model Selection)
Kita memilih algoritma **Neural Network** karena kemampuannya memetakan non-linearitas yang sangat kompleks pada data sekuensial kluster server.

1. **LSTM (Baseline)**: Unggul dalam mengingat dependensi jangka panjang, seperti siklus harian dari workload.
2. **GRU (Gated Recurrent Unit)**: Varian LSTM dengan arsitektur lebih sederhana dan lebih cepat untuk di-training, seringkali performanya setara LSTM.
3. **BiLSTM (Bidirectional LSTM)**: Memproses input dari masa lalu ke masa depan DAN dari masa depan ke masa lalu (menangkap konteks lengkap).
4. **DUCFF (TCN + Transformer)**: *Dual-branch Unified Contextual Feature Fusion*. Arsitektur state-of-the-art yang menggabungkan kemampuan ekstrasi fitur lokal dari TCN (Temporal Convolutional Network) dan ekstraksi dependensi global dari Transformer (Self-Attention mechanism).


```
# 1. Definisi Model LSTM
def build_lstm(input_shape):
    model = Sequential([
        LSTM(64, activation='relu', input_shape=input_shape, return_sequences=False),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(1)
    ], name="LSTM_Model")
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# 2. Definisi Model GRU
def build_gru(input_shape):
    model = Sequential([
        GRU(64, activation='relu', input_shape=input_shape, return_sequences=False),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(1)
    ], name="GRU_Model")
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# 3. Definisi Model BiLSTM
def build_bilstm(input_shape):
    model = Sequential([
        Bidirectional(LSTM(32, activation='relu'), input_shape=input_shape),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(1)
    ], name="BiLSTM_Model")
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

# 4. Definisi Model DUCFF (TCN + Transformer Murni dengan Keras)
def build_ducff(input_shape):
    inputs = Input(shape=input_shape)
    
    # Branch 1: Temporal Convolutional (TCN - Konvolusi 1D)
    x_tcn = tf.keras.layers.Conv1D(filters=32, kernel_size=2, padding='causal', activation='relu')(inputs)
    x_tcn = tf.keras.layers.Conv1D(filters=32, kernel_size=2, padding='causal', dilation_rate=2, activation='relu')(x_tcn)
    x_tcn_flat = tf.keras.layers.Flatten()(x_tcn)
    
    # Branch 2: Transformer (Self-Attention)
    attention_output = tf.keras.layers.MultiHeadAttention(num_heads=2, key_dim=32)(inputs, inputs)
    attention_output = LayerNormalization(epsilon=1e-6)(attention_output + inputs)
    x_transformer_flat = tf.keras.layers.Flatten()(attention_output)
    
    # Fusion (Menggabungkan hasil ekstraksi TCN dan Transformer)
    fusion = tf.keras.layers.Concatenate()([x_tcn_flat, x_transformer_flat])
    x = Dense(64, activation='relu')(fusion)
    x = Dropout(0.2)(x)
    outputs = Dense(1, activation='linear')(x)
    
    model = Model(inputs=inputs, outputs=outputs, name="DUCFF_TCN_Transformer")
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model

input_shape = (TIME_STEPS, len(features))
models = {
    "LSTM": build_lstm(input_shape),
    "GRU": build_gru(input_shape),
    "BiLSTM": build_bilstm(input_shape),
    "DUCFF": build_ducff(input_shape)
}
```

## 2.6 Training (Pelatihan Model)
Model 'belajar' dari data latih (*training set*) dengan menyesuaikan parameter internalnya agar dapat memetakan input (histori workload) ke output (beban CPU masa depan) dengan sebaik mungkin, biasanya dengan meminimalkan fungsi loss (Mean Squared Error).


```
# Membagi Dataset (70% Train, 15% Validation, 15% Test)
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.15, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.176, shuffle=False) # 0.176 dari 0.85 approx 15%

print(f"Train size: {len(X_train)} | Val size: {len(X_val)} | Test size: {len(X_test)}")

# Early Stopping agar training berhenti jika tidak ada peningkatan (mencegah overfit)
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Loop untuk training semua model (Epoch diset rendah 5 untuk contoh eksekusi cepat)
# Secara nyata, epochs = 50 atau 100 dengan batch_size = 64
EPOCHS = 5
BATCH_SIZE = 128

history_dict = {}

for name, model in models.items():
    print(f"\n--- Melatih {name} ---")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stop],
        verbose=1
    )
    history_dict[name] = history
```

## 2.7 Evaluasi Model
Menguji performa model menggunakan data uji (*testing set*) yang belum pernah dilihat model sebelumnya. Kita menggunakan metrik MSE dan MAE.


```
# Menguji ke Test Set
results = {}
predictions = {}

for name, model in models.items():
    test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
    pred = model.predict(X_test, verbose=0)
    
    results[name] = {'MSE': test_loss, 'MAE': test_mae}
    predictions[name] = pred.flatten()

# Tampilkan Hasil dalam bentuk DataFrame
df_results = pd.DataFrame(results).T
print("\n--- Hasil Evaluasi di Test Set ---")
display(df_results)

# Visualisasi Prediksi vs Aktual (Ambil 200 sampel pertama test set agar jelas)
plt.figure(figsize=(15, 6))
plt.plot(y_test[:200], label='Aktual (cpu_media)', color='black', linewidth=2)
for name, pred in predictions.items():
    plt.plot(pred[:200], label=f'Prediksi {name}', alpha=0.8)

plt.title('Perbandingan Prediksi Berbagai Model vs Data Aktual (cpu_media)')
plt.xlabel('Time Step (Sample)')
plt.ylabel('Normalized CPU Load')
plt.legend()
plt.grid(True)
plt.show()
```

## 2.8 Tuning & Optimasi
Jika performa belum memuaskan, dilakukan penyesuaian *hyperparameter* (misalnya *learning rate*, kedalaman *layer*, jumlah neuron) melalui teknik seperti **Grid Search** atau **Random Search**, lalu kembali ke tahap *training*.

*Di bawah ini adalah contoh setup fungsi menggunakan Random Search pada model GRU/LSTM secara manual.*


```
def tune_hyperparameters(X_train, y_train, X_val, y_val):
    # Contoh array hyperparameter
    learning_rates = [0.001, 0.005]
    hidden_units = [32, 64]
    
    best_loss = float('inf')
    best_params = {}
    
    for lr in learning_rates:
        for units in hidden_units:
            print(f"\nMencoba Konfigurasi -> LR: {lr}, Units: {units}")
            # Buat ulang model sementara
            model = Sequential([
                GRU(units, activation='relu', input_shape=(TIME_STEPS, len(features))),
                Dense(1)
            ])
            opt = tf.keras.optimizers.Adam(learning_rate=lr)
            model.compile(optimizer=opt, loss='mse')
            
            # Train singkat
            model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=2, batch_size=256, verbose=0)
            
            # Evaluasi val_loss
            val_loss = model.evaluate(X_val, y_val, verbose=0)
            print(f"Val Loss: {val_loss}")
            
            if val_loss < best_loss:
                best_loss = val_loss
                best_params = {'learning_rate': lr, 'units': units}
                
    return best_params

print("Menjalankan optimasi sederhana...")
# best_params = tune_hyperparameters(X_train, y_train, X_val, y_val)
# print("Best Hyperparameters ditemukan:", best_params)
print("Skipped for execution speed, uncomment the code to run random search.")
```

## Strategi Pembatasan CPU (CPU Cap Strategy)
Berdasarkan hasil prediksi `cpu_media` di masa depan (t+1), kita akan menskalakan *(scale up/down)* jumlah replika. Namun, dalam ekosistem cloud yang dibatasi biayanya, kita harus memiliki **batasan CPU hulu (CPU Limitation)**.

Strateginya:
1. Hitung jumlah replika ideal berdasarkan prediksi CPU.
2. Jika Total CPU yang dibutuhkan = `Jumlah Replika` × `Alokasi CPU per Replika` > `CPU MAX LIMIT CLUSTER`.
3. Maka sistem akan melakukan **Capping (Pembatasan)** jumlah replika maksimum, atau men-*throttle* *traffic*.


```
# Fungsi auto-scaler simulasi dengan limitasi CPU
def calculate_replicas_with_cpu_cap(predicted_cpu_load_normalized, max_cluster_cpu=100.0, cpu_per_replica=20.0):
    '''
    predicted_cpu_load_normalized: Nilai 0-1 hasil neural network
    max_cluster_cpu: Batas limit maksimum persentase CPU di cluster (misal 100%)
    cpu_per_replica: Berapa banyak beban CPU yang mampu ditahan oleh 1 replika (misal 20%)
    '''
    # Kembalikan skala prediksi ke nilai aslinya (Denormalization)
    # Kita buat dummy array sepanjang jumlah kolom, karena minmax scaler mengharap (1, 3) features.
    dummy_input = np.zeros((1, len(df.columns)))
    
    # Anggap kolom indeks ke-0 di data mentah adalah cpu_media
    # Kita cari nilai prediksi aslinya secara sederhana:
    cpu_min = df['cpu_media'].min()
    cpu_max = df['cpu_media'].max()
    predicted_cpu_real = predicted_cpu_load_normalized * (cpu_max - cpu_min) + cpu_min
    
    # 1. Tentukan replika yang dibutuhkan untuk menahan beban ini
    required_replicas = int(np.ceil(predicted_cpu_real / cpu_per_replica))
    if required_replicas < 1:
        required_replicas = 1 # Minimal 1 replika nyala
        
    # 2. STRATEGI PEMBATASAN (CPU LIMITATION)
    # Cek apakah replika x cost per replika melebihi resource MAX cluster
    total_cpu_demand = required_replicas * cpu_per_replica
    
    if total_cpu_demand > max_cluster_cpu:
        capped_replicas = int(np.floor(max_cluster_cpu / cpu_per_replica))
        print(f"[THROTTLING TRIGGERED] Demands {required_replicas} replicas (Total: {total_cpu_demand}% CPU).")
        print(f"-> Capping at {capped_replicas} replicas (Limit {max_cluster_cpu}% CPU).")
        return capped_replicas
    
    return required_replicas

# Contoh Simulasi menggunakan salah satu hasil prediksi model DUCFF
sample_pred_idx = 50
sample_prediction = predictions['DUCFF'][sample_pred_idx]

print(f"Prediksi Neural Network (DUCFF Normalized) t+1: {sample_prediction:.4f}")
final_replicas = calculate_replicas_with_cpu_cap(
    predicted_cpu_load_normalized = sample_prediction,
    max_cluster_cpu = 80.0, # Strategi: limit hanya boleh maksimal memakan 80% total resource cluster
    cpu_per_replica = 25.0  # Misal 1 replika (container) di-set dengan limit 25% CPU usage
)
print(f"Keputusan Auto-scaler: {final_replicas} Replicas")
```
